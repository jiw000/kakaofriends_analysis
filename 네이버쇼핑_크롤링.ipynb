```python
!pip3 install beautifulsoup4 -U
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
answer = []


characters = [
("라이언"),
("카카오무지"),
("카카오콘"),
("어피치"),
("카카오튜브"),
("춘식"),
("카카오네오"),
("카카오프로도"),
("카카오제이지")

]
max_page = 1

for character in characters: 

    for i in tqdm(range(1, max_page+1)):
        url = f"https://search.shopping.naver.com/search/all?frm=NVSHATC&origQuery={character}&pagingIndex={i}&pagingSize=40&productSet=total&query={character}&sort=rel&timestamp=&viewType=list"
        html = requests.get(url).text
        soup = BeautifulSoup(html, "html.parser")
        #div.style_inner__i4gKy

        for n in range(1,41):
            string = f"#__next > div > div.style_container__UxP6u > div > div.style_content_wrap__Cdqnl > div.style_content__xWg5l > ul > div > div:nth-child({n}) > li > div > div.basicList_info_area__TWvzp > div.basicList_title__VfX3c > a"
            stringp = f"#__next > div > div.style_container__UxP6u > div > div.style_content_wrap__Cdqnl > div.style_content__xWg5l > ul > div > div:nth-child({n}) > li > div > div.basicList_info_area__TWvzp > div.basicList_price_area__K7DDT"
            stringk = f"#__next > div > div.style_container__UxP6u > div > div.style_content_wrap__Cdqnl > div.style_content__xWg5l > ul > div > div:nth-child({n}) > li > div > div.basicList_info_area__TWvzp > div.basicList_depth__SbZWF > span:nth-child(3)"
            temp = soup.select(string)
            tempp = soup.select(stringp)
            tempk = soup.select(stringk)
            
            for line in temp:
                for linep in tempp:
                    for linek in tempk:
                        res_detail = {}
                        res_detail['name']=line.text
                        res_detail['price']= linep.text
                        res_detail['keyword']=linek.text
                        res_detail["charater"] = character
                        answer.append(res_detail)
    break
                
answer
```

### 네이버 검색 API를 사용하여 출력(2023.01.22)

```python
import os
import sys
import json
import urllib.request
client_id = "client_id"
client_secret = "secret"

answers = []

querys = [
("라이언"),
("카카오무지"),
("카카오콘"),
("어피치"),
("카카오튜브"),
("춘식"),
("카카오네오"),
("카카오프로도"),
("카카오제이지")

]
for query in querys:
    encText = urllib.parse.quote(query)
    display = "40"

    for page in range(1,410,40):
        url = "https://openapi.naver.com/v1/search/shop?query=" + encText + "&display=" + display + "&start=" + str(page)
        request = urllib.request.Request(url)
        request.add_header("X-Naver-Client-Id",client_id)
        request.add_header("X-Naver-Client-Secret",client_secret)
        response = urllib.request.urlopen(request)
        rescode = response.getcode()
        if(rescode==200):
            response_body = response.read()
       # print(response_body.decode('utf-8'))
            a = response_body.decode('utf-8')
            b= json.loads(a)
            answers.append(b)
            
        else:
            print("Error Code:" + rescode)
    
       
answers
```

```python
name = []
price =[]
category = []

for c in range(0,90):
    for i in range(0,40):
        name.append(answers[c]['items'][i]["title"])
        price.append(answers[c]['items'][i]["lprice"])
        category.append(answers[c]['items'][i]["category3"])
```

```python
result=pd.DataFrame({"name":name,
                    "price":price,
                    "category":category,
                    })
result["name"] = result["name"].replace('<b>',"  ")
result
```

```python
result["character"] = "무지"
result["character"][0:400] = "라이언"
result["character"][801:1200] = "콘"
result["character"][1201:1600] = "어피치"
result["character"][1601:2000] = "튜브"
result["character"][2001:2400] = "춘식"
result["character"][2401:2800] = "네오"
result["character"][2801:3200] = "프로도"
result["character"][3201:3600] = "제이지"



result
```
